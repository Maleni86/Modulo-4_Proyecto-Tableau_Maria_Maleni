``Limpieza:``

**Columna date_added**

- Hacer dos columnas. Una para mes/año.
- Modificar columna a tipo int.

```python
# Valores que alberga la columna date_added. Es una columna de tipo object

df_final["date_added"]
```

```python
# Dividimos la columna en 3 nuevas columnas (mes, día, año)

split_cols = df_final['date_added'].str.split('/', expand=True) # expand=True hace que el resultado se guarde en columnas separadas

# Asignamos el mes y el año a las nuevas columnas que hemos creado y las convertimos a int

df_final['month_added'] = split_cols[0].astype(int) # split_cols[0] → Sobre la separación, cogemos la columna del mes   .astype(int) → convierte esos valores que están como texto a números
df_final['year_added'] = split_cols[2].astype(int)  # split_cols[2] → Sobre la separación, cogemos la columna del año   
```

```python
# Comprobamos que se han añadido las columnas month_added y year_added (al final del df)

df_final.head()
```

```python
# Borramos la columna original (date_added) que ya no aporta valor

df_final = df_final.drop(columns=['date_added'])
```
