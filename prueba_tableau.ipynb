# Función para limpiar la columna 'year'

def clean_year(year):
    # Eliminar los paréntesis
    year = year.replace('(', '').replace(')', '')
    
    # Eliminar cualquier texto adicional (como "I", "II", etc.) si está presente
    year = year.replace('II', '').replace('I', '').strip()
    
    # Si hay un guion, tomar el primer año antes del guion
    if '–' in year:
        return year.split('–')[0]
    return year  # Si no hay guion, devolvemos el año como está

# Aplicar la limpieza
df['year_cleaned'] = df['year'].apply(clean_year)

print(df[['year', 'year_cleaned']])
